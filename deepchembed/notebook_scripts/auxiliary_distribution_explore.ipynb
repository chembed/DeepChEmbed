```python
import __init__
import matplotlib.pyplot as plt
import numpy as np
from dce import DCE
```

```python
a_0 = np.random.random(size=100)
a_1 = np.random.random(size=100)*(1-a_0)
q_2 = np.array([a_0, 1-a_0]).T
q_3 = np.array([a_0, a_1, 1-a_0-a_1]).T
```

## Explore old target disctribution

```python
hardening_func = DCE.simple_hardening
p_n_1 = hardening_func(q_2, n=1)
p_n_2 = hardening_func(q_2, n=2)
p_n_3 = hardening_func(q_2, n=3)
p_n_4 = hardening_func(q_2, n=4)
p_n_5 = hardening_func(q_2, n=5)

plt.figure(figsize=(10,8))
plt.plot(q_2.T[0], p_n_1.T[0], 'o', label='n=1.00')
plt.plot(q_2.T[0], p_n_2.T[0], 'o', label='n=2.00')
plt.plot(q_2.T[0], p_n_3.T[0], 'o', label='n=3.00')
plt.plot(q_2.T[0], p_n_4.T[0], 'o', label='n=4.00')
plt.plot(q_2.T[0], p_n_5.T[0], 'o', label='n=5.00')
plt.axhline(0.5, c='k')
plt.plot([0,1],[0,1],c='k')
plt.xlabel('Soft Label')
plt.ylabel('Hardend Label')
plt.legend()
```

```python
hardening_func = DCE.simple_hardening
p_n_1 = hardening_func(q_3, n=1)
p_n_2 = hardening_func(q_3, n=2)
p_n_3 = hardening_func(q_3, n=3)
p_n_4 = hardening_func(q_3, n=4)
p_n_5 = hardening_func(q_3, n=5)

plt.figure(figsize=(10,8))
plt.plot(q_3.T[0], p_n_1.T[0], 'o', label='n=1.00')
plt.plot(q_3.T[0], p_n_2.T[0], 'o', label='n=2.00')
plt.plot(q_3.T[0], p_n_3.T[0], 'o', label='n=3.00')
plt.plot(q_3.T[0], p_n_4.T[0], 'o', label='n=4.00')
plt.plot(q_3.T[0], p_n_5.T[0], 'o', label='n=5.00')
plt.plot([0,1],[0,1],c='k')
plt.axhline(0.333333, c='k')
plt.axhline(0.666667, c='k')
plt.legend()
```

## Explore smoothstep target disctribution

```python
hardening_func = DCE.smoothstep_hardening
p_n_1 = hardening_func(q_2, n=1)
p_n_2 = hardening_func(q_2, n=3)
p_n_3 = hardening_func(q_2, n=5)
p_n_4 = hardening_func(q_2, n=7)
p_n_5 = hardening_func(q_2, n=9)

plt.figure(figsize=(10,8))
plt.plot(q_2.T[0], p_n_1.T[0], 'o', label='n=1.00')
plt.plot(q_2.T[0], p_n_2.T[0], 'o', label='n=3.00')
plt.plot(q_2.T[0], p_n_3.T[0], 'o', label='n=5.00')
plt.plot(q_2.T[0], p_n_4.T[0], 'o', label='n=7.00')
plt.plot(q_2.T[0], p_n_5.T[0], 'o', label='n=9.00')
plt.axhline(0.5, c='k')
plt.plot([0,1],[0,1],c='k')
plt.xlabel('Soft Label')
plt.ylabel('Hardend Label')
plt.legend()
```
