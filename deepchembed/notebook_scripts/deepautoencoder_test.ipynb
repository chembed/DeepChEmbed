```python
import __init__
from __init__ import DATA_PATH
from __init__ import PACKAGE_PATH
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
data = pd.read_csv(os.path.join(DATA_PATH,'BOD_biclass_139rdkit.csv'))
data_139 = data.iloc[:,2:141]
data_train = data_139.values
data_train.shape
```

```python
from dimreducer import DeepAutoEncoder
```

## Single layer

```python
dims = [139,35]
autoencoder = DeepAutoEncoder(dims)
```

```python
test_dims = [10, 20, 35, 50, 70, 100]
train_histories = []
input_dim = 139

for dim in test_dims:
    autoencoder.set_dims([input_dim, dim])
    autoencoder.build_autoencoder()
    history = autoencoder.train_model(data_train)
    train_histories.append(history.history['loss'])
```

```python
for i in range(len(test_dims)):
    plt.plot(train_histories[i], label = 'embeded_dim = '+str(test_dims[i]))

plt.legend()
```

```python
converged_loss = []
for i in range(len(test_dims)):
    converged_loss.append(train_histories[i][-1])

plt.plot(test_dims, converged_loss)
```
